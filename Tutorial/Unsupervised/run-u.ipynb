```python
import subprocess
import sys
import time

def run_and_log(command, log_file_name):
    """
    ËøêË°åÂëΩ‰ª§ÔºåÂπ∂Â∞ÜËæìÂá∫ÂêåÊó∂ÊòæÁ§∫Âú® Jupyter Â±èÂπï‰∏äÂíå‰øùÂ≠òÂà∞Êó•ÂøóÊñá‰ª∂‰∏≠„ÄÇ
    """
    print(f"üöÄ ÂºÄÂßãËøêË°å: {command}")
    print(f"üìÅ Êó•ÂøóÂ∞Ü‰øùÂ≠òÂú®: {log_file_name}")
    print("-" * 50)
    
    # ‰ª•ËøΩÂä†Ê®°ÂºèÊâìÂºÄÊó•ÂøóÊñá‰ª∂ÔºåÈò≤Ê≠¢Ë¶ÜÁõñ‰πãÂâçÁöÑËÆ∞ÂΩï
    with open(log_file_name, "w", encoding="utf-8") as f:
        # ÂêØÂä®ËøõÁ®ã
        process = subprocess.Popen(
            command, 
            stdout=subprocess.PIPE, 
            stderr=subprocess.STDOUT, # ÊääÈîôËØØ‰ø°ÊÅØ‰πüÂêàÂπ∂Âà∞ËæìÂá∫Èáå
            shell=True,
            text=True, # ‰ª•ÊñáÊú¨ÂΩ¢ÂºèËØªÂèñ
            encoding='utf-8', # Èò≤Ê≠¢‰∏≠Êñá‰π±Á†ÅÔºåÂ¶ÇÊûúÊä•ÈîôÊîπÊàê 'gbk'
            bufsize=1 # Ë°åÁºìÂÜ≤
        )
        
        # ÂÆûÊó∂ËØªÂèñËæìÂá∫
        while True:
            line = process.stdout.readline()
            if not line and process.poll() is not None:
                break
            if line:
                # 1. ÊâìÂç∞Âà∞Â±èÂπï (Jupyter)
                sys.stdout.write(line)
                sys.stdout.flush() 
                # 2. ÂÜôÂÖ•Âà∞Êñá‰ª∂
                f.write(line)
                f.flush()
                
    if process.returncode == 0:
        print("-" * 50)
        print("‚úÖ ËøêË°åÊàêÂäüÂÆåÊàêÔºÅ")
    else:
        print("-" * 50)
        print("‚ùå ËøêË°åÂá∫ÈîôÔºåËØ∑Ê£ÄÊü•‰∏äÊñπÁöÑÊä•Èîô‰ø°ÊÅØ„ÄÇ")
```
